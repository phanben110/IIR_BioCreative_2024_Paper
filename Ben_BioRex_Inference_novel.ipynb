```python
import pandas as pd
import os
relation_path = "/content/drive/MyDrive/BioCreativeVIII/BioREx/tmp/final_result.pubtator"
# Initialize data collection list
list_data_NER = []
list_data_NER_raw = []
list_data_RE = []
text_t_in_documents = {}
text_a_in_documents = {}

# Open and process the file
with open(relation_path, 'r') as file:

    for line in file:
        fields = line.strip().split('\t')
        if len(fields) > 1:
            for i in range(len(fields)):
                if "MESH:" in fields[i]:
                  fields[i] = fields[i][5:]
                #   print(field)

        if len(fields) == 6:

            list_data_NER_raw.append([fields[0],fields[1], fields[2], fields[3], fields[4], fields[5]]) 
            list_id_identifier = fields[5].split(',')  
            if len(fields[5].split(',')) > 1: 
                for identifier in list_id_identifier: 
                    list_data_NER.append([fields[0],fields[1], fields[2], fields[3], fields[4], identifier]) 
            else: 
                list_data_NER.append([fields[0],fields[1], fields[2], fields[3], fields[4], fields[5]])
        elif len(fields) == 1 and '|t|' in fields[0]:
            position_t = fields[0].find('|t|')
            id_document = fields[0][:position_t]
            text_t_in_documents.update({id_document :fields[0][position_t+3:]})
        elif len(fields) == 1 and '|a|' in fields[0]:
            position_a = fields[0].find('|a|')
            id_document = fields[0][:position_a]
            text_a_in_documents.update({id_document :fields[0][position_a+3:]})
        elif len(fields) == 5:
            list_data_RE.append([fields[0],fields[1], fields[2], fields[3], fields[4]])
data_frame_RE = pd.DataFrame(list_data_RE, columns=["doc_id", "relation", "id_identifier_1", "id_identifier_2", "acc_relation"])
data_frame_NER = pd.DataFrame(list_data_NER, columns=["doc_id", "offset_1", "offset_2", "entity", "identifier", "id_identifier"])
data_frame_NER_raw = pd.DataFrame(list_data_NER_raw, columns=["doc_id", "offset_1", "offset_2", "entity", "identifier", "id_identifier"])
```

```python
import torch
from tqdm import tqdm
from transformers import BertTokenizer, BertModel
import os
from datetime import datetime
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import warnings
warnings.filterwarnings("ignore")

# set the maximum sequence length
max_len = 100
batch_size = 2048
name_bert = "alvaroalon2/biobert_diseases_ner"
device = torch.device("cuda")
num_classes = 2

# name_bert = "microsoft/BiomedNLP-PubMedBERT-large-uncased-abstract"
```

```python
# Tokenize and encode the sentences
class MyDataset(Dataset):
    def __init__(self, df, tokenizer, max_len):
        self.data = df
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        sentence1 = self.data.iloc[index]['full_sentence_1']
        sentence2 = self.data.iloc[index]['full_sentence_2']
        e1 = self.data.iloc[index]['entity_1']
        e2 = self.data.iloc[index]['entity_2']
        identifier1 = self.data.iloc[index]['identifier_1']
        identifier2 = self.data.iloc[index]['identifier_2']
        relation = self.data.iloc[index]['relation']


        # novel_cls = self.data.iloc[index]['novel_cls']
        # input_text = f"{e1} [SEP] {e2}"
        #input_text = f"{e1} - {identifier1} - {sentence1} [SEP] {e2} - {identifier2} - {sentence2} [SEP] {relation}"
        input_text = f"{e1} - {identifier1} - {sentence1} [SEP] {e2} - {identifier2} - {sentence2} [SEP] {relation}"

        encoding = self.tokenizer.encode_plus(
            input_text,
            add_special_tokens=True,
            truncation=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'text': input_text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten()
        }
```

```python
# # create an instance of the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained(name_bert)
# create data loaders for the training and validation sets
infer_data = MyDataset(merged_df_2, tokenizer, max_len)
infer_loader = DataLoader(infer_data, batch_size=batch_size, shuffle=True)
```

```python
class BertSentimentClassifier(nn.Module):
    def __init__(self, bert_model_name, num_classes):
        super(BertSentimentClassifier, self).__init__()

        self.bert = BertModel.from_pretrained(bert_model_name)
        self.dropout1 = nn.Dropout(0.2)
        self.fc1 = nn.Linear(self.bert.config.hidden_size, 256)
        self.dropout2 = nn.Dropout(0.2)
        self.fc2 = nn.Linear(256, num_classes)


    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        pooled_output = self.dropout1(pooled_output)
        output = self.fc1(pooled_output)
        output = self.dropout2(output)
        logits = self.fc2(output)

        return logits
```

```python
def inferences(model, data_loader,device):
    # set the model to eval mode
    model.eval()

    # iterate over the data loader
    for data in tqdm(data_loader):
        # move the inputs to the device
        input_ids = data['input_ids'].to(device)
        attention_mask = data['attention_mask'].to(device)

        # disable gradient computation
        with torch.no_grad():
            # get the model's predictions
            outputs = model(input_ids, attention_mask)
            # calculate the number of correct predictions
            logits = outputs.detach().cpu().numpy()

    return logits
```

```python
def load_bert_model(weight_path, bert_model_name, device, num_classes ):

    model_predict = BertSentimentClassifier(bert_model_name, num_classes)
    #model_predict.load_state_dict(torch.load(weight_path, map_location=torch.device('gpu')))
    model_predict.load_state_dict(torch.load(weight_path, map_location=device))
    model_predict.to(device)
    return model_predict

weight_path = "/content/drive/MyDrive/BioCreativeVIII/BioREx/models/Novel_cls_task/Model_BioMeidcal_small/biobert_disease_ner_v1.pt"
model_predict = load_bert_model(weight_path, name_bert, device, num_classes)
```

```python
# set the model to eval mode
model_predict.eval()
import numpy as np
# Create an empty list to store pred_flat outputs
all_predictions = []

# iterate over the data loader
for data in tqdm(infer_loader):
    # move the inputs to the device
    input_ids = data['input_ids'].to(device)
    attention_mask = data['attention_mask'].to(device)

    # disable gradient computation
    with torch.no_grad():
        outputs = model_predict(input_ids, attention_mask)
        logits = outputs.detach().cpu().numpy()
        pred_flat = np.argmax(logits, axis=1).flatten()

        # Append pred_flat to the list
        all_predictions.append(pred_flat)

# Convert the list of arrays to a single NumPy array
all_predictions = np.concatenate(all_predictions, axis=0)

# Print or use the all_predictions array as needed
print("All Predictions: ", all_predictions)
np.save('/content/drive/MyDrive/BioCreativeVIII/BioREx/tmp/all_predictions.npy',all_predictions)
```

```python
def predict_aveg_cls_novel(row): 
    try:
        doc_id = row["doc_id"]
        relation = row["relation"]
        id_identifier_1 = row["id_identifier_1"]
        id_identifier_2 = row["id_identifier_2"]

        vote = final_dataframe[(final_dataframe["doc_id"] == doc_id) &
                        (final_dataframe["relation"] == relation) &
                        (final_dataframe["id_identifier_1"] == id_identifier_1) &
                        (final_dataframe["id_identifier_2"] == id_identifier_2)]["cls_novel"].value_counts()
      #   print(vote)

        if len(vote) == 1: 
           return str(vote.index[0])
        else:
          if vote[0] == vote[1] :
             return "1"
          elif vote[0] > vote[1] : 
             return "0"
          elif vote[1] > vote[0]: 
             return "1"
    except Exception as e:
        print(f"Error processing row: {row}")
        print(f"Error details: {e}")
        # You can choose to return a default value or handle the error in another way
        return None
```

```python
def extract_list_NER(doc_id): 
    data_frame = data_frame_NER_raw[["doc_id"] == doc_id]
    list_NER = []
    for index, row in data_frame.iterrows(): 
        ner_row = row["doc_id"] + "\t"  + row['offset_1'] + "\t" + row['offset_2'] + "\t" + row['entity'] + "\t" + row['identifier'] + "\t" + row['id_identifier'] 
        list_NER.append(ner_row)
    return list_NER
```

```python
def extract_list_RE(doc_id): 
    data_frame = data_frame_RE_threshold[data_frame_RE_threshold["doc_id"] == doc_id]
    list_RE = []
    for index, row in data_frame.iterrows(): 
        ner_row = row["doc_id"] + "\t"  + row['relation'] + "\t" + row['id_identifier_1'] + "\t" + row['id_identifier_2'] + "\t" + row['cls_novel']
        list_RE.append(ner_row)
    return list_RE
```
